```python

df = pd.read_csv('train.csv')#
df.head(n=5)
```

```python
print(df1.head(n=5))
print(df1.mean())
print(ma.sqrt(df1.var()))
```

```python
mu=df1.mean()
sd=ma.sqrt(df1.var())
x=df1
# the histogram of the data
n, bins, patches = hist(x, 50, normed=1, alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sd)
l = plt.plot(bins, y, 'r--', linewidth=1,color = "b")

plt.xlabel('Elevation')
plt.ylabel('Probability')
plt.title("Histogram of elevation forest cover type 1")
plt.axis([2200, 3900, 0, 0.003])
plt.grid(True)

plt.show()
```
